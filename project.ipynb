```python
strbks_df = pd.read_csv(url)
# strbks_df.head()
```

```python
strbks_df.columns
```

#### keys:
+ gender:

     **0**: Male; **1**: Female
+ age: 

     **0**: Below 20; **1**: 20-29; **2**: 30-39; **3**: 40 and above
+ status:

     **0**: Student; **1**: Self-Employed; **2**: Employed; **3**: Housewife
+ income (in Malaysian Ringgit):

     **0**: Less than 25,000; **1**: 25,00050,000; **2**: 50,000100,000; **3**: 100,000150,000; **4**: More than 150,000
+ visitNo:

     **0**: Daily; **1**: Weekly; **3**: Monthly; **4**: Never
